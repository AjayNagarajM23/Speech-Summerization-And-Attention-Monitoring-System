```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('yolov5', 'custom', path='best.pt', force_reload=True , source='local')
```

```python
img = 'Images/DSC_0006.JPG'
results = model(img,)
print(results)
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture(1)
while True:
    ret, frame = cap.read()

    # Make detections
    results = model(frame)

    cv2.imshow('YOLO', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
crops = results.crop(save=True, save_dir='Images/', hidelabels=True)
```
